```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import text

# Connection parameters
DB_USER = "rpLesson15"
DB_PASSWORD = "Republic$01"
DB_HOST = "mysql"     # Docker Compose service name for MySQL
DB_PORT = 3306
DB_NAME = "rp_schema"

# Correct connection string using pymysql

connection_url = f"mysql+pymysql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"

engine = create_engine(connection_url, pool_pre_ping=True)
# Test connection


# Establish a connection
with engine.connect() as conn:
    result = conn.execute(text("SELECT * FROM student"))
    rows = result.fetchall()
    columns = result.keys()

# Load into pandas manually
df_student = pd.DataFrame(rows, columns=columns)
display(df_student)
```
