```python
import pandas as pd
import os
from pathlib import Path

# Define the relative directory containing your CSV files
csv_dir = Path("data")

# Dictionary to hold DataFrames: key = filename (without .csv), value = DataFrame
dataframes = {}

# Loop through all files in the directory
for filename in os.listdir(csv_dir):
    if filename.endswith(".csv"): #retrive files that end with .csv
        file_path = csv_dir / filename
        df_name = filename.replace(".csv", "")
        df = pd.read_csv(file_path)
        dataframes[df_name] = df
        print(f"âœ… Loaded {filename} into dataframes['{df_name}']")

# Optional: display one of them
# display(dataframes['your_file_name_without_extension'].head())
# Display each DataFrame with its name
for name, df in dataframes.items():
    print(f"\nðŸ“„ DataFrame: '{name}' â€” {len(df)} rows Ã— {len(df.columns)} columns")
    display(df.head())  # Shows only the first 5 rows of each DataFrame
```
